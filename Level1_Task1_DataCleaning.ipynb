```python
import pandas as pd
import numpy as np

df = pd.read_csv('../Level1 Task2_EDA/Amazon Sale Report.csv', low_memory=False)

print("Shape:", df.shape)
print("\nFirst look:")
df.head()
```

```python
# ── 2. Inspect the Data ──────────────────────────────────────
print("Column names:\n", df.columns.tolist())
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
```

```python
# ── 3. Handle Missing Values ──────────────────────────────────

# Drop columns with more than 50% missing or useless columns
df.drop(columns=['fulfilled-by', 'Unnamed: 22', 'promotion-ids'], inplace=True)
print("Columns after dropping high-missing cols:", df.shape)

# Fill Courier Status with mode
df['Courier Status'].fillna(df['Courier Status'].mode()[0], inplace=True)

# Fill currency with mode
df['currency'].fillna(df['currency'].mode()[0], inplace=True)

# Fill Amount with median
df['Amount'].fillna(df['Amount'].median(), inplace=True)

# Fill ship location columns with 'Unknown'
for col in ['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']:
    df[col].fillna('Unknown', inplace=True)

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# ── 3. Handle Missing Values ──────────────────────────────────

# Drop columns with more than 50% missing or useless columns
df.drop(columns=['fulfilled-by', 'Unnamed: 22', 'promotion-ids'], inplace=True)
print("Columns after dropping high-missing cols:", df.shape)

# Fill Courier Status with mode
df['Courier Status'] = df['Courier Status'].fillna(df['Courier Status'].mode()[0])

# Fill currency with mode
df['currency'] = df['currency'].fillna(df['currency'].mode()[0])

# Fill Amount with median
df['Amount'] = df['Amount'].fillna(df['Amount'].median())

# Fill ship location columns with 0 (they are numeric)
for col in ['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']:
    df[col] = df[col].fillna('Unknown').astype(str)

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# ── 3. Handle Missing Values ──────────────────────────────────

# Drop columns with more than 50% missing (use errors='ignore' in case already dropped)
cols_to_drop = ['fulfilled-by', 'Unnamed: 22', 'promotion-ids']
df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])
print("Columns after dropping:", df.shape)

# Fill Courier Status with mode
df['Courier Status'] = df['Courier Status'].fillna(df['Courier Status'].mode()[0])

# Fill currency with mode
df['currency'] = df['currency'].fillna(df['currency'].mode()[0])

# Fill Amount with median
df['Amount'] = df['Amount'].fillna(df['Amount'].median())

# Fill ship location columns
for col in ['ship-city', 'ship-state', 'ship-postal-code', 'ship-country']:
    df[col] = df[col].astype(str).fillna('Unknown')

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# ── 4. Remove Duplicates ──────────────────────────────────────
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
print("New shape:", df.shape)
```

```python
# ── 5. Standardize Data Formats ──────────────────────────────
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Standardize categorical columns
cat_cols = ['Status', 'Fulfilment', 'Sales Channel', 'ship-service-level',
            'Category', 'Size', 'Courier Status']

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

print("Date sample:", df['Date'].head())
print("\nStatus values:", df['Status'].unique())
print("\nCategory values:", df['Category'].unique())
```

```python
# ── 6. Save Cleaned Dataset ──────────────────────────────────
df.to_csv('Amazon_Sales_Cleaned.csv', index=False)
print("Cleaned dataset saved.")
print("Final shape:", df.shape)
df.head()
```
